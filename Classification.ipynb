Decision tree 

```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
X, y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.plot_tree(clf)
```

```python
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("iris") 
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, 
                     feature_names=iris.feature_names,  
                     class_names=iris.target_names,  
                     filled=True, rounded=True,  
                     special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

Sampling a training set while holding out 40% of the data for testing (evaluating) our classifier:

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=0)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
clf = clf.fit(X, y)
clf.score(X_test, y_test)
```

Using cross-validation by calling the cross_val_score function on the estimator and the dataset.

The following example demonstrates how to estimate the accuracy of a decision tree on the iris dataset by splitting the data, fitting a model and computing the score 5 consecutive times (with different splits each time):

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, X, y, cv=5)
```

The mean score and the standard deviation are hence given by:

```python
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
from sklearn.model_selection import cross_validate
scores = cross_validate(
    clf, X, y, cv=5, scoring='f1_macro')
scores
```

Running cross-validation on multiple metrics and also to return train scores, fit times and score times.

```python
scoring = ['precision_macro', 'recall_macro','f1_macro']
from sklearn.model_selection import cross_validate
scores = cross_validate(
    clf, X, y, cv=5, scoring=scoring)
scores
```
